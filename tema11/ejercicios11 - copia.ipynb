````python
# Prueba tu sistema - Parte 3: Reservaciones
print("\n=== RESERVACIONES ===")

# Crear reservaciones
print("\nCreando reservaciones...")
hoy = datetime.now().date()
manana = hoy + timedelta(days=1)

res1 = restaurante.crear_reservacion(
    'Carlos Ruiz', 
    '555-1234',
    f'{manana} 20:00:00',
    4
)

res2 = restaurante.crear_reservacion(
    'Laura S√°nchez',
    '555-5678',
    f'{manana} 21:00:00',
    2
)

res3 = restaurante.crear_reservacion(
    'Pedro G√≥mez',
    '555-9012',
    f'{manana} 19:30:00',
    6
)

print(f"‚úÖ Reservaci√≥n {res1} creada")
print(f"‚úÖ Reservaci√≥n {res2} creada")
print(f"‚úÖ Reservaci√≥n {res3} creada")

# Confirmar reservaciones
print("\nConfirmando reservaciones...")
restaurante.confirmar_reservacion(res1, 5)  # Mesa 5 para 4 personas
restaurante.confirmar_reservacion(res2, 1)  # Mesa 1 para 2 personas
restaurante.confirmar_reservacion(res3, 9)  # Mesa 9 para 6 personas

# Ver reservaciones del d√≠a
print(f"\nReservaciones para {manana}:")
reservaciones = restaurante.reservaciones_del_dia(manana)
for res in reservaciones:
    print(f"  {res['fecha_hora']} - {res['nombre_cliente']} ({res['num_personas']} personas) - Mesa {res['numero_mesa']} - {res['estado']}")
````

````python
# Prueba tu sistema - Parte 4: Reportes y Estad√≠sticas
print("\n=== REPORTES Y ESTAD√çSTICAS ===")

# Cerrar las √≥rdenes restantes para tener datos completos
restaurante.cerrar_orden(orden2_id)
restaurante.cerrar_orden(orden3_id)

# Ventas del d√≠a
print("\nVentas de hoy:")
ventas = restaurante.ventas_del_dia()
print(f"  Total de √≥rdenes: {ventas['num_ordenes']}")
print(f"  Total vendido: ${ventas['total']:.2f}")
print(f"  Promedio por orden: ${ventas['promedio']:.2f}")

# Platillos m√°s vendidos
print("\nTop 5 platillos m√°s vendidos:")
top_platillos = restaurante.platillos_mas_vendidos(5)
for platillo in top_platillos:
    print(f"  {platillo['nombre']:<25} {platillo['cantidad_vendida']} unidades - ${platillo['ingresos']:.2f}")

# Promedio de consumo por mesa
print("\nPromedio de consumo por mesa:")
promedio = restaurante.promedio_consumo_por_mesa()
print(f"  ${promedio:.2f}")

# Reporte completo
print("\n=== REPORTE COMPLETO ===")
fecha_inicio = datetime.now().date()
fecha_fin = fecha_inicio
reporte = restaurante.reporte_completo(fecha_inicio, fecha_fin)

print(f"\nPer√≠odo: {fecha_inicio} - {fecha_fin}")
print(f"\nVentas:")
print(f"  √ìrdenes totales: {reporte['ordenes_totales']}")
print(f"  Ingresos totales: ${reporte['ingresos_totales']:.2f}")
print(f"  Promedio por orden: ${reporte['promedio_por_orden']:.2f}")

print(f"\nProductos:")
print(f"  Items vendidos: {reporte['items_vendidos']}")
print(f"  Platillo m√°s popular: {reporte['platillo_mas_popular']}")

print(f"\nReservaciones:")
print(f"  Total: {reporte['reservaciones_totales']}")
print(f"  Confirmadas: {reporte['reservaciones_confirmadas']}")
print(f"  Canceladas: {reporte['reservaciones_canceladas']}")
````
