- Tenga un l√≠mite de elementos en cach√© (LRU - Least Recently Used)

Usa `collections.OrderedDict`.

**Ejemplo:**
```python
@cache_resultado(max_size=3)
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
```

````python
# Prueba tu decorador
import time

@cache_resultado(max_size=5)
def operacion_costosa(n):
    """Simula una operaci√≥n que toma tiempo."""
    time.sleep(0.5)  # Simular trabajo
    return n ** 2

print("Primera llamada (deber√≠a calcular):")
print(f"Resultado: {operacion_costosa(5)}\n")

print("Segunda llamada con mismo argumento (deber√≠a usar cach√©):")
print(f"Resultado: {operacion_costosa(5)}\n")

print("Llamadas con diferentes argumentos:")
for i in range(1, 8):
    print(f"operacion_costosa({i}) = {operacion_costosa(i)}")
````
