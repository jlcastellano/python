### 5.4. Script Python para Firmar AutomÃ¡ticamente

````python
# Script para automatizar la firma de ejecutables

script_firma = '''
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Script de Firma AutomÃ¡tica
Firma ejecutables usando SignTool de Windows
"""

import os
import subprocess
import sys
from pathlib import Path

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CONFIGURACIÃ“N - EDITAR ESTOS VALORES
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Ruta al certificado .pfx
CERTIFICADO = r"C:\\Certificados\\mi_certificado.pfx"

# ContraseÃ±a del certificado (mejor usar variable de entorno)
PASSWORD = os.environ.get('CERT_PASSWORD', 'tu_contraseÃ±a_aqui')

# Ruta a SignTool.exe
SIGNTOOL = r"C:\\Program Files (x86)\\Windows Kits\\10\\bin\\10.0.22621.0\\x64\\signtool.exe"

# Timestamp server
TIMESTAMP = "http://timestamp.digicert.com"

# InformaciÃ³n del software
DESCRIPCION = "Mi AplicaciÃ³n"
URL = "https://www.miapp.com"

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def verificar_requisitos():
    """Verifica que existan los archivos necesarios."""
    errores = []
    
    if not os.path.exists(SIGNTOOL):
        errores.append(f"âŒ No se encuentra SignTool en: {SIGNTOOL}")
    
    if not os.path.exists(CERTIFICADO):
        errores.append(f"âŒ No se encuentra el certificado en: {CERTIFICADO}")
    
    if errores:
        for error in errores:
            print(error)
        return False
    
    return True

def firmar_archivo(archivo_exe):
    """
    Firma un archivo ejecutable con firma dual (SHA-1 + SHA-256).
    
    Args:
        archivo_exe: Ruta al archivo .exe a firmar
    
    Returns:
        bool: True si la firma fue exitosa
    """
    if not os.path.exists(archivo_exe):
        print(f"âŒ Archivo no encontrado: {archivo_exe}")
        return False
    
    print(f"\n{'='*70}")
    print(f"  FIRMANDO: {os.path.basename(archivo_exe)}")
    print(f"{'='*70}\n")
    
    # Primera firma: SHA-1 (compatibilidad con Windows 7)
    print("ğŸ“ Aplicando firma SHA-1...")
    comando_sha1 = [
        SIGNTOOL,
        "sign",
        "/f", CERTIFICADO,
        "/p", PASSWORD,
        "/d", DESCRIPCION,
        "/du", URL,
        "/t", TIMESTAMP,
        "/v",  # Verbose
        archivo_exe
    ]
    
    try:
        resultado = subprocess.run(
            comando_sha1,
            capture_output=True,
            text=True,
            check=True
        )
        print("âœ… Firma SHA-1 aplicada correctamente")
    except subprocess.CalledProcessError as e:
        print(f"âŒ Error en firma SHA-1:\n{e.stderr}")
        return False
    
    # Segunda firma: SHA-256 (recomendado para Windows 8+)
    print("\nğŸ“ Aplicando firma SHA-256...")
    comando_sha256 = [
        SIGNTOOL,
        "sign",
        "/f", CERTIFICADO,
        "/p", PASSWORD,
        "/fd", "sha256",
        "/tr", TIMESTAMP,
        "/td", "sha256",
        "/as",  # Append signature
        "/v",
        archivo_exe
    ]
    
    try:
        resultado = subprocess.run(
            comando_sha256,
            capture_output=True,
            text=True,
            check=True
        )
        print("âœ… Firma SHA-256 aplicada correctamente")
    except subprocess.CalledProcessError as e:
        print(f"âŒ Error en firma SHA-256:\n{e.stderr}")
        return False
    
    # Verificar firma
    print("\nğŸ” Verificando firma...")
    verificar = [
        SIGNTOOL,
        "verify",
        "/pa",
        "/v",
        archivo_exe
    ]
    
    try:
        resultado = subprocess.run(
            verificar,
            capture_output=True,
            text=True,
            check=True
        )
        print("âœ… Firma verificada correctamente")
        print(f"\n{resultado.stdout}")
        return True
    except subprocess.CalledProcessError as e:
        print(f"âŒ Error en verificaciÃ³n:\n{e.stderr}")
        return False

def firmar_directorio(directorio, patron="*.exe"):
    """
    Firma todos los ejecutables en un directorio.
    
    Args:
        directorio: Ruta al directorio
        patron: PatrÃ³n de archivos a firmar (default: *.exe)
    """
    archivos = list(Path(directorio).glob(patron))
    
    if not archivos:
        print(f"âŒ No se encontraron archivos {patron} en {directorio}")
        return
    
    print(f"\nğŸ“¦ Encontrados {len(archivos)} archivo(s) para firmar\n")
    
    exitosos = 0
    fallidos = 0
    
    for archivo in archivos:
        if firmar_archivo(str(archivo)):
            exitosos += 1
        else:
            fallidos += 1
    
    print(f"\n{'='*70}")
    print(f"  RESUMEN")
    print(f"{'='*70}")
    print(f"âœ… Exitosos: {exitosos}")
    print(f"âŒ Fallidos:  {fallidos}")
    print(f"{'='*70}\n")

def main():
    """FunciÃ³n principal."""
    print("\n" + "="*70)
    print("  SCRIPT DE FIRMA DIGITAL DE EJECUTABLES")
    print("="*70 + "\n")
    
    # Verificar requisitos
    if not verificar_requisitos():
        print("\nâŒ Verifica la configuraciÃ³n y vuelve a intentar")
        sys.exit(1)
    
    print("âœ… VerificaciÃ³n de requisitos completada\n")
    
    # Ejemplo de uso
    if len(sys.argv) > 1:
        # Si se pasa un argumento, firmar ese archivo/directorio
        ruta = sys.argv[1]
        if os.path.isfile(ruta):
            firmar_archivo(ruta)
        elif os.path.isdir(ruta):
            firmar_directorio(ruta)
        else:
            print(f"âŒ Ruta no vÃ¡lida: {ruta}")
    else:
        # Firmar archivos en la carpeta dist/
        if os.path.exists("dist"):
            firmar_directorio("dist")
        else:
            print("âŒ No se encontrÃ³ carpeta 'dist'")
            print("\nUso: python firmar.py [archivo.exe | directorio]")

if __name__ == "__main__":
    main()
'''

with open('firmar.py', 'w', encoding='utf-8') as f:
    f.write(script_firma)

print("âœ… Script 'firmar.py' creado")
print("\nğŸ“ Antes de usar:")
print("   1. Edita las variables de configuraciÃ³n al inicio del script")
print("   2. Configura CERT_PASSWORD como variable de entorno")
print("   3. Verifica la ruta de SignTool.exe")
print("\nğŸš€ Uso:")
print("   python firmar.py archivo.exe")
print("   python firmar.py dist/")
````

## 8. IntegraciÃ³n con PyInstaller

````python
# Script completo: compilar y firmar automÃ¡ticamente

script_completo = '''
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Script Completo: Compilar y Firmar
Compila con PyInstaller y firma el ejecutable automÃ¡ticamente
"""

import os
import sys
import subprocess
import shutil
from pathlib import Path

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CONFIGURACIÃ“N
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# PyInstaller
SCRIPT_PYTHON = "mi_app.py"
NOMBRE_APP = "MiAplicacion"
ICONO = "resources/icon.ico"

# Firma digital (Windows)
SIGNTOOL = r"C:\\Program Files (x86)\\Windows Kits\\10\\bin\\10.0.22621.0\\x64\\signtool.exe"
CERTIFICADO = r"C:\\Certificados\\certificado.pfx"
PASSWORD = os.environ.get('CERT_PASSWORD', '')
TIMESTAMP = "http://timestamp.digicert.com"

# InformaciÃ³n
VERSION = "1.0.0"
DESCRIPCION = "Mi AplicaciÃ³n IncreÃ­ble"
URL = "https://www.miapp.com"

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def limpiar():
    """Limpia directorios de builds anteriores."""
    print("ğŸ§¹ Limpiando directorios...")
    for directorio in ["build", "dist"]:
        if os.path.exists(directorio):
            shutil.rmtree(directorio)
            print(f"   âœ“ {directorio}/ eliminado")

def compilar():
    """Compila con PyInstaller."""
    print(f"\nğŸ”¨ Compilando {NOMBRE_APP}...\n")
    
    comando = [
        "pyinstaller",
        "--onefile",
        "--noconsole",
        f"--name={NOMBRE_APP}",
        f"--icon={ICONO}",
        "--clean",
        SCRIPT_PYTHON
    ]
    
    try:
        subprocess.run(comando, check=True)
        print("\nâœ… CompilaciÃ³n exitosa")
        return True
    except subprocess.CalledProcessError as e:
        print(f"\nâŒ Error en compilaciÃ³n: {e}")
        return False

def firmar_windows(ejecutable):
    """Firma el ejecutable en Windows."""
    if not all([os.path.exists(SIGNTOOL), os.path.exists(CERTIFICADO), PASSWORD]):
        print("âš ï¸  ConfiguraciÃ³n de firma incompleta, omitiendo firma...")
        return True
    
    print(f"\nâœï¸  Firmando {os.path.basename(ejecutable)}...")
    
    # Firma SHA-1
    cmd_sha1 = [
        SIGNTOOL, "sign",
        "/f", CERTIFICADO,
        "/p", PASSWORD,
        "/d", DESCRIPCION,
        "/du", URL,
        "/t", TIMESTAMP,
        ejecutable
    ]
    
    try:
        subprocess.run(cmd_sha1, check=True, capture_output=True)
        print("   âœ“ Firma SHA-1 aplicada")
    except subprocess.CalledProcessError:
        print("   âŒ Error en firma SHA-1")
        return False
    
    # Firma SHA-256
    cmd_sha256 = [
        SIGNTOOL, "sign",
        "/f", CERTIFICADO,
        "/p", PASSWORD,
        "/fd", "sha256",
        "/tr", TIMESTAMP,
        "/td", "sha256",
        "/as",
        ejecutable
    ]
    
    try:
        subprocess.run(cmd_sha256, check=True, capture_output=True)
        print("   âœ“ Firma SHA-256 aplicada")
    except subprocess.CalledProcessError:
        print("   âŒ Error en firma SHA-256")
        return False
    
    # Verificar
    cmd_verify = [SIGNTOOL, "verify", "/pa", ejecutable]
    try:
        subprocess.run(cmd_verify, check=True, capture_output=True)
        print("   âœ“ Firma verificada correctamente")
        return True
    except subprocess.CalledProcessError:
        print("   âŒ Error en verificaciÃ³n")
        return False

def crear_zip():
    """Crea archivo ZIP del ejecutable."""
    print("\nğŸ“¦ Creando archivo ZIP...")
    
    extension = ".exe" if sys.platform == "win32" else ""
    ejecutable = Path("dist") / f"{NOMBRE_APP}{extension}"
    zip_nombre = f"{NOMBRE_APP}_v{VERSION}_{sys.platform}"
    
    shutil.make_archive(
        os.path.join("dist", zip_nombre),
        'zip',
        'dist',
        ejecutable.name
    )
    
    print(f"   âœ“ {zip_nombre}.zip creado")

def generar_hashes():
    """Genera hashes del ejecutable."""
    import hashlib
    
    print("\nğŸ” Generando hashes...")
    
    extension = ".exe" if sys.platform == "win32" else """
    ejecutable = Path("dist") / f"{NOMBRE_APP}{extension}"
    
    with open(ejecutable, 'rb') as f:
        contenido = f.read()
        md5 = hashlib.md5(contenido).hexdigest()
        sha256 = hashlib.sha256(contenido).hexdigest()
    
    hash_file = Path("dist") / f"{NOMBRE_APP}_hashes.txt"
    with open(hash_file, 'w') as f:
        f.write(f"{NOMBRE_APP} v{VERSION}\n")
        f.write(f"{'='*60}\n\n")
        f.write(f"MD5:    {md5}\n")
        f.write(f"SHA256: {sha256}\n")
    
    print(f"   MD5:    {md5}")
    print(f"   SHA256: {sha256}")
    print(f"   âœ“ Guardado en {hash_file}")

def main():
    """FunciÃ³n principal."""
    print("\n" + "="*70)
    print(f"  COMPILACIÃ“N Y FIRMA DE {NOMBRE_APP.upper()} v{VERSION}")
    print("="*70)
    
    # Verificar script principal
    if not os.path.exists(SCRIPT_PYTHON):
        print(f"\nâŒ No se encontrÃ³ {SCRIPT_PYTHON}")
        sys.exit(1)
    
    # Proceso completo
    limpiar()
    
    if not compilar():
        sys.exit(1)
    
    # Firmar si es Windows
    if sys.platform == "win32":
        ejecutable = Path("dist") / f"{NOMBRE_APP}.exe"
        if ejecutable.exists():
            firmar_windows(str(ejecutable))
    
    # Crear ZIP y generar hashes
    crear_zip()
    generar_hashes()
    
    print("\n" + "="*70)
    print("  âœ… PROCESO COMPLETADO EXITOSAMENTE")
    print("="*70)
    print(f"\nğŸ“‚ Archivos en dist/:")
    for archivo in Path("dist").glob("*"):
        tamaÃ±o = archivo.stat().st_size / (1024 * 1024)
        print(f"   â€¢ {archivo.name} ({tamaÃ±o:.2f} MB)")
    print()

if __name__ == "__main__":
    main()
'''

with open('compilar_y_firmar.py', 'w', encoding='utf-8') as f:
    f.write(script_completo)

print("âœ… Script 'compilar_y_firmar.py' creado")
print("\nğŸ“ ConfiguraciÃ³n necesaria:")
print("   1. Edita las rutas en la secciÃ³n CONFIGURACIÃ“N")
print("   2. Configura variable de entorno CERT_PASSWORD")
print("   3. Ejecuta: python compilar_y_firmar.py")
````
