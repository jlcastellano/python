## 1. Â¿QuÃ© es un Ejecutable?

Un **ejecutable** es un archivo que puede ejecutarse directamente sin necesidad de tener Python instalado.

### Ventajas de crear ejecutables:
- **DistribuciÃ³n fÃ¡cil:** Los usuarios no necesitan instalar Python
- **ProtecciÃ³n del cÃ³digo:** El cÃ³digo fuente no es directamente visible
- **Profesionalismo:** Apariencia mÃ¡s profesional para aplicaciones
- **Simplicidad:** Doble clic para ejecutar
- **InstalaciÃ³n simple:** No requiere dependencias externas

### Desventajas:
- **TamaÃ±o:** Los ejecutables suelen ser grandes (incluyen el intÃ©rprete)
- **EspecÃ­ficos de plataforma:** Necesitas compilar para cada SO
- **Tiempo de inicio:** Pueden tardar mÃ¡s en arrancar
- **Actualizaciones:** Requiere redistribuir el ejecutable completo

### Herramientas principales:
- **PyInstaller** â­ (mÃ¡s popular y completo)
- **Auto-py-to-exe** (interfaz grÃ¡fica para PyInstaller)
- **cx_Freeze** (alternativa robusta)
- **Nuitka** (compila a C++, mÃ¡s rÃ¡pido)
- **PyOxidizer** (mÃ¡s moderno, de Mozilla)

## 2. PyInstaller - La Herramienta Principal

**PyInstaller** es la herramienta mÃ¡s popular y versÃ¡til para crear ejecutables de Python.

### CaracterÃ­sticas:
- Compatible con Windows, macOS y Linux
- Soporta la mayorÃ­a de bibliotecas de Python
- Puede crear ejecutables de un solo archivo
- Detecta automÃ¡ticamente dependencias
- Soporta GUI (Tkinter, PyQt, etc.)

### 2.1. InstalaciÃ³n

````python
# InstalaciÃ³n de PyInstaller
# Ejecuta en tu terminal o CMD:

# pip install pyinstaller

# Verificar instalaciÃ³n
# pyinstaller --version

# En este notebook, verificamos si estÃ¡ instalado
import subprocess
import sys

try:
    result = subprocess.run(
        [sys.executable, "-m", "pip", "show", "pyinstaller"],
        capture_output=True,
        text=True
    )
    if result.returncode == 0:
        print("âœ… PyInstaller estÃ¡ instalado")
        print(result.stdout)
    else:
        print("âŒ PyInstaller no estÃ¡ instalado")
        print("Ejecuta: pip install pyinstaller")
except Exception as e:
    print(f"Error al verificar: {e}")
````

### 2.2. Comandos BÃ¡sicos de PyInstaller

#### Comando mÃ¡s simple:
```bash
pyinstaller mi_script.py
```

#### Opciones comunes:
```bash
# Crear un solo archivo ejecutable
pyinstaller --onefile mi_script.py

# Eliminar la consola (solo GUI)
pyinstaller --onefile --noconsole mi_script.py

# Agregar un icono
pyinstaller --onefile --icon=icono.ico mi_script.py

# Especificar nombre del ejecutable
pyinstaller --onefile --name MiApp mi_script.py

# Agregar archivos de datos
pyinstaller --onefile --add-data "data.txt;." mi_script.py
```

### 2.3. Opciones Importantes

| OpciÃ³n | DescripciÃ³n |
|--------|-------------|
| `--onefile` | Crea un solo archivo ejecutable |
| `--onedir` | Crea un directorio con el ejecutable (por defecto) |
| `--noconsole` | No muestra ventana de consola (Windows) |
| `--windowed` | Igual que `--noconsole` |
| `--console` | Muestra ventana de consola (por defecto) |
| `--icon=archivo.ico` | Agrega icono al ejecutable |
| `--name NOMBRE` | Nombre del ejecutable |
| `--add-data SRC;DEST` | Incluye archivos de datos |
| `--add-binary SRC;DEST` | Incluye archivos binarios |
| `--hidden-import MODULO` | Incluye mÃ³dulos no detectados |
| `--exclude-module MODULO` | Excluye mÃ³dulos innecesarios |
| `--clean` | Limpia cachÃ© antes de compilar |
| `--log-level LEVEL` | Nivel de detalle (DEBUG, INFO, WARN) |

### 3.1. Script Simple de Consola

````python
# Crear script: calculadora_simple.py

script_calculadora = '''
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Calculadora Simple
Una calculadora de lÃ­nea de comandos
"""

def sumar(a, b):
    """Suma dos nÃºmeros."""
    return a + b

def restar(a, b):
    """Resta dos nÃºmeros."""
    return a - b

def multiplicar(a, b):
    """Multiplica dos nÃºmeros."""
    return a * b

def dividir(a, b):
    """Divide dos nÃºmeros."""
    if b == 0:
        return "Error: DivisiÃ³n por cero"
    return a / b

def main():
    """FunciÃ³n principal."""
    print("="*50)
    print("  CALCULADORA SIMPLE")
    print("="*50)
    
    operaciones = {
        '1': ('Sumar', sumar),
        '2': ('Restar', restar),
        '3': ('Multiplicar', multiplicar),
        '4': ('Dividir', dividir)
    }
    
    while True:
        print("\\nOperaciones disponibles:")
        for key, (nombre, _) in operaciones.items():
            print(f"  {key}. {nombre}")
        print("  0. Salir")
        
        opcion = input("\\nSelecciona una operaciÃ³n: ")
        
        if opcion == '0':
            print("Â¡Hasta luego!")
            break
        
        if opcion not in operaciones:
            print("âŒ OpciÃ³n invÃ¡lida")
            continue
        
        try:
            a = float(input("Primer nÃºmero: "))
            b = float(input("Segundo nÃºmero: "))
            
            nombre, funcion = operaciones[opcion]
            resultado = funcion(a, b)
            print(f"\\nâœ… Resultado: {resultado}")
            
        except ValueError:
            print("âŒ Error: Debes ingresar nÃºmeros vÃ¡lidos")
        except Exception as e:
            print(f"âŒ Error: {e}")
        
        input("\\nPresiona Enter para continuar...")

if __name__ == "__main__":
    main()
'''

# Guardar el script
with open('calculadora_simple.py', 'w', encoding='utf-8') as f:
    f.write(script_calculadora)

print("âœ… Archivo 'calculadora_simple.py' creado")
print("\nPara crear el ejecutable, ejecuta en tu terminal:")
print("pyinstaller --onefile --name Calculadora calculadora_simple.py")
````

### 3.2. Script con Interfaz GrÃ¡fica (Tkinter)

````python
# Crear script: calculadora_gui.py

script_gui = '''
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Calculadora GUI
Una calculadora con interfaz grÃ¡fica usando Tkinter
"""

import tkinter as tk
from tkinter import messagebox

class CalculadoraGUI:
    def __init__(self, master):
        self.master = master
        master.title("Calculadora")
        master.geometry("400x500")
        master.resizable(False, False)
        
        # Variable para mostrar resultado
        self.resultado = tk.StringVar()
        self.resultado.set("0")
        
        # Variables de cÃ¡lculo
        self.operacion = None
        self.primer_numero = None
        
        self.crear_interfaz()
    
    def crear_interfaz(self):
        """Crea la interfaz de la calculadora."""
        # Display
        display = tk.Entry(
            self.master,
            textvariable=self.resultado,
            font=('Arial', 24),
            bd=10,
            justify='right',
            state='readonly'
        )
        display.grid(row=0, column=0, columnspan=4, padx=10, pady=10)
        
        # Botones
        botones = [
            ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
            ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
            ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
            ('0', 4, 0), ('.', 4, 1), ('=', 4, 2), ('+', 4, 3),
            ('C', 5, 0), ('CE', 5, 1), ('â†', 5, 2)
        ]
        
        for (texto, fila, col) in botones:
            if texto == '=':
                cmd = self.calcular
                bg = '#4CAF50'
            elif texto in ['C', 'CE']:
                cmd = self.limpiar if texto == 'C' else self.limpiar_entrada
                bg = '#f44336'
            elif texto == 'â†':
                cmd = self.borrar_ultimo
                bg = '#FF9800'
            elif texto in ['+', '-', '*', '/']:
                cmd = lambda t=texto: self.operador(t)
                bg = '#2196F3'
            else:
                cmd = lambda t=texto: self.agregar_numero(t)
                bg = '#9E9E9E'
            
            span = 2 if texto == '0' else 1
            btn = tk.Button(
                self.master,
                text=texto,
                font=('Arial', 18),
                bg=bg,
                fg='white',
                command=cmd,
                width=3 if span == 1 else 7,
                height=2
            )
            btn.grid(row=fila, column=col, columnspan=span, padx=5, pady=5)
    
    def agregar_numero(self, numero):
        """Agrega un nÃºmero al display."""
        actual = self.resultado.get()
        if actual == "0":
            self.resultado.set(numero)
        else:
            self.resultado.set(actual + numero)
    
    def operador(self, op):
        """Guarda la operaciÃ³n a realizar."""
        try:
            self.primer_numero = float(self.resultado.get())
            self.operacion = op
            self.resultado.set("0")
        except ValueError:
            messagebox.showerror("Error", "NÃºmero invÃ¡lido")
    
    def calcular(self):
        """Realiza el cÃ¡lculo."""
        if self.primer_numero is None or self.operacion is None:
            return
        
        try:
            segundo_numero = float(self.resultado.get())
            
            if self.operacion == '+':
                resultado = self.primer_numero + segundo_numero
            elif self.operacion == '-':
                resultado = self.primer_numero - segundo_numero
            elif self.operacion == '*':
                resultado = self.primer_numero * segundo_numero
            elif self.operacion == '/':
                if segundo_numero == 0:
                    messagebox.showerror("Error", "DivisiÃ³n por cero")
                    self.limpiar()
                    return
                resultado = self.primer_numero / segundo_numero
            
            self.resultado.set(str(resultado))
            self.primer_numero = None
            self.operacion = None
            
        except ValueError:
            messagebox.showerror("Error", "NÃºmero invÃ¡lido")
    
    def limpiar(self):
        """Limpia todo."""
        self.resultado.set("0")
        self.primer_numero = None
        self.operacion = None
    
    def limpiar_entrada(self):
        """Limpia solo la entrada actual."""
        self.resultado.set("0")
    
    def borrar_ultimo(self):
        """Borra el Ãºltimo carÃ¡cter."""
        actual = self.resultado.get()
        if len(actual) > 1:
            self.resultado.set(actual[:-1])
        else:
            self.resultado.set("0")

def main():
    root = tk.Tk()
    app = CalculadoraGUI(root)
    root.mainloop()

if __name__ == "__main__":
    main()
'''

# Guardar el script
with open('calculadora_gui.py', 'w', encoding='utf-8') as f:
    f.write(script_gui)

print("âœ… Archivo 'calculadora_gui.py' creado")
print("\nPara crear el ejecutable, ejecuta en tu terminal:")
print("pyinstaller --onefile --noconsole --name CalculadoraGUI calculadora_gui.py")
````

### 3.3. Script con Recursos Externos

````python
# Crear un archivo de datos de ejemplo
datos_json = '''
{
    "app_name": "Mi AplicaciÃ³n",
    "version": "1.0.0",
    "autor": "Tu Nombre",
    "mensajes": {
        "bienvenida": "Â¡Bienvenido a la aplicaciÃ³n!",
        "despedida": "Â¡Hasta luego!"
    }
}
'''

with open('config.json', 'w', encoding='utf-8') as f:
    f.write(datos_json)

# Script que usa recursos externos
script_recursos = '''
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
AplicaciÃ³n con Recursos Externos
Demuestra cÃ³mo incluir archivos de datos en el ejecutable
"""

import json
import os
import sys

def obtener_ruta_recurso(nombre_archivo):
    """
    Obtiene la ruta correcta del archivo de recursos.
    Funciona tanto en desarrollo como en el ejecutable.
    """
    try:
        # PyInstaller crea una carpeta temporal y guarda la ruta en _MEIPASS
        ruta_base = sys._MEIPASS
    except Exception:
        # En desarrollo, usa el directorio actual
        ruta_base = os.path.abspath(".")
    
    return os.path.join(ruta_base, nombre_archivo)

def cargar_configuracion():
    """Carga el archivo de configuraciÃ³n."""
    try:
        ruta = obtener_ruta_recurso('config.json')
        with open(ruta, 'r', encoding='utf-8') as f:
            return json.load(f)
    except FileNotFoundError:
        print("âŒ Error: No se encontrÃ³ el archivo de configuraciÃ³n")
        return None
    except json.JSONDecodeError:
        print("âŒ Error: El archivo de configuraciÃ³n no es vÃ¡lido")
        return None

def main():
    """FunciÃ³n principal."""
    config = cargar_configuracion()
    
    if config is None:
        return
    
    print("="*60)
    print(f"  {config['app_name']} v{config['version']}")
    print(f"  Autor: {config['autor']}")
    print("="*60)
    print(f"\\n{config['mensajes']['bienvenida']}")
    
    print("\\nContenido de la configuraciÃ³n:")
    print(json.dumps(config, indent=2, ensure_ascii=False))
    
    input(f"\\n{config['mensajes']['despedida']}\\nPresiona Enter para salir...")

if __name__ == "__main__":
    main()
'''

with open('app_con_recursos.py', 'w', encoding='utf-8') as f:
    f.write(script_recursos)

print("âœ… Archivos creados:")
print("   - config.json")
print("   - app_con_recursos.py")
print("\\nPara crear el ejecutable, ejecuta:")
print("pyinstaller --onefile --add-data 'config.json;.' app_con_recursos.py")
print("\\nEn Linux/Mac usa ':' en lugar de ';':")
print("pyinstaller --onefile --add-data 'config.json:.' app_con_recursos.py")
````

## 4. Archivo de EspecificaciÃ³n (.spec)

PyInstaller genera un archivo `.spec` que puedes editar para personalizar la compilaciÃ³n.

### Ventajas del archivo .spec:
- Mayor control sobre el proceso de compilaciÃ³n
- Reutilizable y versionable
- Permite configuraciones complejas

````python
# Ejemplo de archivo .spec personalizado

spec_ejemplo = '''
# -*- mode: python ; coding: utf-8 -*-

block_cipher = None

a = Analysis(
    ['calculadora_gui.py'],  # Script principal
    pathex=[],
    binaries=[],
    datas=[('config.json', '.')],  # Archivos de datos
    hiddenimports=[],  # MÃ³dulos ocultos
    hookspath=[],
    hooksconfig={},
    runtime_hooks=[],
    excludes=['matplotlib', 'numpy'],  # Excluir mÃ³dulos grandes
    win_no_prefer_redirects=False,
    win_private_assemblies=False,
    cipher=block_cipher,
    noarchive=False,
)

pyz = PYZ(a.pure, a.zipped_data, cipher=block_cipher)

exe = EXE(
    pyz,
    a.scripts,
    a.binaries,
    a.zipfiles,
    a.datas,
    [],
    name='MiCalculadora',  # Nombre del ejecutable
    debug=False,
    bootloader_ignore_signals=False,
    strip=False,
    upx=True,  # Comprimir con UPX
    upx_exclude=[],
    runtime_tmpdir=None,
    console=False,  # Sin consola
    disable_windowed_traceback=False,
    argv_emulation=False,
    target_arch=None,
    codesign_identity=None,
    entitlements_file=None
)
'''

with open('ejemplo.spec', 'w', encoding='utf-8') as f:
    f.write(spec_ejemplo)

print("âœ… Archivo 'ejemplo.spec' creado")
print("\nPara compilar usando el archivo .spec:")
print("pyinstaller ejemplo.spec")
print("\nPara generar un archivo .spec sin compilar:")
print("pyi-makespec --onefile --noconsole mi_script.py")
````

### 8.2. Script de ConstrucciÃ³n Automatizado

````python
# Crear script de construcciÃ³n automatizado

script_build = '''
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Script de ConstrucciÃ³n Automatizado
Automatiza el proceso de crear ejecutables
"""

import os
import shutil
import subprocess
import sys
from pathlib import Path

# ConfiguraciÃ³n
SCRIPT_PRINCIPAL = "src/main.py"
NOMBRE_APP = "MiAplicacion"
ICONO = "resources/icons/app.ico"
VERSION = "1.0.0"

# Archivos de datos a incluir
DATOS = [
    ("resources/data/config.json", "data"),
    ("resources/images", "images"),
]

def limpiar_directorios():
    """Limpia directorios de construcciÃ³n anteriores."""
    print("ğŸ§¹ Limpiando directorios...")
    for directorio in ["build", "dist"]:
        if os.path.exists(directorio):
            shutil.rmtree(directorio)
            print(f"   âœ“ {directorio}/ eliminado")

def construir_comando():
    """Construye el comando de PyInstaller."""
    comando = [
        "pyinstaller",
        "--onefile",
        "--noconsole",
        f"--name={NOMBRE_APP}",
        f"--icon={ICONO}",
        "--clean",
    ]
    
    # Agregar archivos de datos
    for origen, destino in DATOS:
        separador = ";" if sys.platform == "win32" else ":"
        comando.append(f"--add-data={origen}{separador}{destino}")
    
    comando.append(SCRIPT_PRINCIPAL)
    
    return comando

def ejecutar_pyinstaller():
    """Ejecuta PyInstaller."""
    print(f"\nğŸ”¨ Construyendo {NOMBRE_APP} v{VERSION}...")
    
    comando = construir_comando()
    print(f"\nComando: {' '.join(comando)}\n")
    
    try:
        resultado = subprocess.run(
            comando,
            check=True,
            capture_output=True,
            text=True
        )
        print(resultado.stdout)
        return True
    except subprocess.CalledProcessError as e:
        print(f"âŒ Error durante la construcciÃ³n:")
        print(e.stderr)
        return False

def verificar_ejecutable():
    """Verifica que el ejecutable fue creado."""
    extension = ".exe" if sys.platform == "win32" else ""
    ejecutable = Path("dist") / f"{NOMBRE_APP}{extension}"
    
    if ejecutable.exists():
        tamaÃ±o_mb = ejecutable.stat().st_size / (1024 * 1024)
        print(f"\nâœ… Ejecutable creado exitosamente!")
        print(f"   ğŸ“ UbicaciÃ³n: {ejecutable.absolute()}")
        print(f"   ğŸ“Š TamaÃ±o: {tamaÃ±o_mb:.2f} MB")
        return True
    else:
        print(f"\nâŒ No se encontrÃ³ el ejecutable en {ejecutable}")
        return False

def crear_zip():
    """Crea un archivo ZIP del ejecutable."""
    print(f"\nğŸ“¦ Creando archivo ZIP...")
    
    extension = ".exe" if sys.platform == "win32" else ""
    ejecutable = Path("dist") / f"{NOMBRE_APP}{extension}"
    zip_nombre = f"{NOMBRE_APP}_v{VERSION}_{sys.platform}"
    
    shutil.make_archive(
        os.path.join("dist", zip_nombre),
        'zip',
        'dist',
        ejecutable.name
    )
    
    print(f"   âœ“ {zip_nombre}.zip creado")

def main():
    """FunciÃ³n principal."""
    print("="*60)
    print(f"  CONSTRUCCIÃ“N DE {NOMBRE_APP.upper()} v{VERSION}")
    print("="*60)
    
    # Verificar que existe el script principal
    if not os.path.exists(SCRIPT_PRINCIPAL):
        print(f"âŒ Error: No se encontrÃ³ {SCRIPT_PRINCIPAL}")
        sys.exit(1)
    
    # Proceso de construcciÃ³n
    limpiar_directorios()
    
    if not ejecutar_pyinstaller():
        print("\nâŒ La construcciÃ³n fallÃ³")
        sys.exit(1)
    
    if not verificar_ejecutable():
        sys.exit(1)
    
    # Crear ZIP
    crear_zip()
    
    print("\n" + "="*60)
    print("  âœ… Â¡CONSTRUCCIÃ“N COMPLETADA EXITOSAMENTE!")
    print("="*60)

if __name__ == "__main__":
    main()
'''

with open('build.py', 'w', encoding='utf-8') as f:
    f.write(script_build)

print("âœ… Archivo 'build.py' creado")
print("\nPara construir tu aplicaciÃ³n, ejecuta:")
print("python build.py")
````

## 10. Ejemplos de Aplicaciones Completas

### 10.1. AplicaciÃ³n GUI Completa

````python
# AplicaciÃ³n de gestiÃ³n de tareas completa

app_completa = '''
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Gestor de Tareas
AplicaciÃ³n completa de gestiÃ³n de tareas con GUI
"""

import tkinter as tk
from tkinter import ttk, messagebox
import json
import os
import sys
from datetime import datetime

class GestorTareas:
    def __init__(self, master):
        self.master = master
        self.master.title("Gestor de Tareas v1.0")
        self.master.geometry("800x600")
        
        self.tareas = []
        self.archivo_datos = self.obtener_ruta_datos()
        
        self.crear_interfaz()
        self.cargar_tareas()
    
    def obtener_ruta_datos(self):
        """Obtiene la ruta del archivo de datos."""
        if getattr(sys, 'frozen', False):
            # Ejecutable
            ruta_base = os.path.dirname(sys.executable)
        else:
            # Script
            ruta_base = os.path.dirname(os.path.abspath(__file__))
        
        return os.path.join(ruta_base, 'tareas.json')
    
    def crear_interfaz(self):
        """Crea la interfaz de usuario."""
        # Frame superior
        frame_top = ttk.Frame(self.master, padding="10")
        frame_top.pack(fill=tk.X)
        
        ttk.Label(frame_top, text="Nueva tarea:").pack(side=tk.LEFT)
        
        self.entry_tarea = ttk.Entry(frame_top, width=50)
        self.entry_tarea.pack(side=tk.LEFT, padx=5)
        self.entry_tarea.bind('<Return>', lambda e: self.agregar_tarea())
        
        ttk.Button(
            frame_top,
            text="Agregar",
            command=self.agregar_tarea
        ).pack(side=tk.LEFT)
        
        # Frame de lista
        frame_lista = ttk.Frame(self.master, padding="10")
        frame_lista.pack(fill=tk.BOTH, expand=True)
        
        # Scrollbar
        scrollbar = ttk.Scrollbar(frame_lista)
        scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
        
        # Treeview
        self.tree = ttk.Treeview(
            frame_lista,
            columns=('Estado', 'Tarea', 'Fecha'),
            show='headings',
            yscrollcommand=scrollbar.set
        )
        
        self.tree.heading('Estado', text='Estado')
        self.tree.heading('Tarea', text='Tarea')
        self.tree.heading('Fecha', text='Fecha')
        
        self.tree.column('Estado', width=100)
        self.tree.column('Tarea', width=500)
        self.tree.column('Fecha', width=150)
        
        self.tree.pack(fill=tk.BOTH, expand=True)
        scrollbar.config(command=self.tree.yview)
        
        # Frame de botones
        frame_botones = ttk.Frame(self.master, padding="10")
        frame_botones.pack(fill=tk.X)
        
        ttk.Button(
            frame_botones,
            text="Completar",
            command=self.completar_tarea
        ).pack(side=tk.LEFT, padx=5)
        
        ttk.Button(
            frame_botones,
            text="Eliminar",
            command=self.eliminar_tarea
        ).pack(side=tk.LEFT, padx=5)
        
        ttk.Button(
            frame_botones,
            text="Limpiar Completadas",
            command=self.limpiar_completadas
        ).pack(side=tk.LEFT, padx=5)
    
    def agregar_tarea(self):
        """Agrega una nueva tarea."""
        texto = self.entry_tarea.get().strip()
        
        if not texto:
            messagebox.showwarning("Advertencia", "Ingresa una tarea")
            return
        
        tarea = {
            'texto': texto,
            'completada': False,
            'fecha': datetime.now().strftime('%Y-%m-%d %H:%M')
        }
        
        self.tareas.append(tarea)
        self.actualizar_lista()
        self.guardar_tareas()
        self.entry_tarea.delete(0, tk.END)
    
    def completar_tarea(self):
        """Marca una tarea como completada."""
        seleccion = self.tree.selection()
        
        if not seleccion:
            messagebox.showwarning("Advertencia", "Selecciona una tarea")
            return
        
        indice = self.tree.index(seleccion[0])
        self.tareas[indice]['completada'] = True
        self.actualizar_lista()
        self.guardar_tareas()
    
    def eliminar_tarea(self):
        """Elimina la tarea seleccionada."""
        seleccion = self.tree.selection()
        
        if not seleccion:
            messagebox.showwarning("Advertencia", "Selecciona una tarea")
            return
        
        if messagebox.askyesno("Confirmar", "Â¿Eliminar esta tarea?"):
            indice = self.tree.index(seleccion[0])
            del self.tareas[indice]
            self.actualizar_lista()
            self.guardar_tareas()
    
    def limpiar_completadas(self):
        """Elimina todas las tareas completadas."""
        completadas = sum(1 for t in self.tareas if t['completada'])
        
        if completadas == 0:
            messagebox.showinfo("Info", "No hay tareas completadas")
            return
        
        if messagebox.askyesno(
            "Confirmar",
            f"Â¿Eliminar {completadas} tarea(s) completada(s)?"
        ):
            self.tareas = [t for t in self.tareas if not t['completada']]
            self.actualizar_lista()
            self.guardar_tareas()
    
    def actualizar_lista(self):
        """Actualiza la lista de tareas en la interfaz."""
        # Limpiar
        for item in self.tree.get_children():
            self.tree.delete(item)
        
        # Agregar tareas
        for tarea in self.tareas:
            estado = "âœ… Completada" if tarea['completada'] else "â³ Pendiente"
            self.tree.insert(
                '',
                tk.END,
                values=(estado, tarea['texto'], tarea['fecha'])
            )
    
    def cargar_tareas(self):
        """Carga las tareas desde el archivo."""
        try:
            if os.path.exists(self.archivo_datos):
                with open(self.archivo_datos, 'r', encoding='utf-8') as f:
                    self.tareas = json.load(f)
                self.actualizar_lista()
        except Exception as e:
            messagebox.showerror("Error", f"Error al cargar tareas: {e}")
    
    def guardar_tareas(self):
        """Guarda las tareas en el archivo."""
        try:
            with open(self.archivo_datos, 'w', encoding='utf-8') as f:
                json.dump(self.tareas, f, indent=2, ensure_ascii=False)
        except Exception as e:
            messagebox.showerror("Error", f"Error al guardar tareas: {e}")

def main():
    root = tk.Tk()
    app = GestorTareas(root)
    root.mainloop()

if __name__ == "__main__":
    main()
'''

with open('gestor_tareas.py', 'w', encoding='utf-8') as f:
    f.write(app_completa)

print("âœ… Archivo 'gestor_tareas.py' creado")
print("\nPara crear el ejecutable:")
print("pyinstaller --onefile --noconsole --name GestorTareas gestor_tareas.py")
````
