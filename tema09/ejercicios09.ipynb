### Ejercicio 6: Procesador de Datos CSV üü°

Crea una funci√≥n `procesar_csv(archivo_entrada, archivo_salida, transformaciones)` que:
- Lea un archivo CSV
- Aplique transformaciones a las columnas especificadas
- Guarde el resultado en un nuevo archivo CSV

Las transformaciones son un diccionario donde:
- La clave es el nombre de la columna
- El valor es una funci√≥n que transforma el valor

**Ejemplo:**
```python
transformaciones = {
    'nombre': str.upper,
    'edad': lambda x: int(x) + 1,
    'salario': lambda x: float(x) * 1.1
}
procesar_csv('empleados.csv', 'empleados_actualizado.csv', transformaciones)
```

````python
# Tu c√≥digo aqu√≠
````

````python
# Prueba tu funci√≥n
import csv

# Crear archivo CSV de prueba
with open('empleados.csv', 'w', newline='', encoding='utf-8') as f:
    escritor = csv.DictWriter(f, fieldnames=['nombre', 'edad', 'salario'])
    escritor.writeheader()
    escritor.writerow({'nombre': 'ana', 'edad': '25', 'salario': '2000'})
    escritor.writerow({'nombre': 'luis', 'edad': '30', 'salario': '2500'})
    escritor.writerow({'nombre': 'maria', 'edad': '28', 'salario': '2200'})

# Aplicar transformaciones
transformaciones = {
    'nombre': str.upper,
    'edad': lambda x: int(x) + 1,
    'salario': lambda x: f"{float(x) * 1.1:.2f}"
}

procesar_csv('empleados.csv', 'empleados_actualizado.csv', transformaciones)

# Mostrar resultado
print("Empleados actualizados:")
with open('empleados_actualizado.csv', 'r', encoding='utf-8') as f:
    lector = csv.DictReader(f)
    for fila in lector:
        print(f"  {fila}")
````

````python
# Prueba tu sistema
analizador = AnalizadorLogs()
analizador.cargar_archivo('sistema.log')

print("=== AN√ÅLISIS DE LOGS ===")
print(f"Total de entradas: {len(analizador.logs)}\n")

# Contar por nivel
print("Distribuci√≥n por nivel:")
por_nivel = analizador.contar_por_nivel()
for nivel, cantidad in por_nivel.items():
    print(f"  {nivel}: {cantidad}")

# Contar por usuario
print("\nActividad por usuario:")
por_usuario = analizador.contar_por_usuario()
for usuario, cantidad in sorted(por_usuario.items(), key=lambda x: x[1], reverse=True):
    print(f"  {usuario}: {cantidad} acciones")

# Contar por acci√≥n
print("\nAcciones m√°s comunes:")
por_accion = analizador.contar_por_accion()
for accion, cantidad in sorted(por_accion.items(), key=lambda x: x[1], reverse=True)[:5]:
    print(f"  {accion}: {cantidad}")

# Detectar intentos fallidos
print("\nDetecci√≥n de m√∫ltiples intentos fallidos:")
sospechosos = analizador.detectar_intentos_fallidos(max_intentos=2, ventana_minutos=10)
for usuario, info in sospechosos.items():
    print(f"  ‚ö†Ô∏è  {usuario}: {info['intentos']} intentos fallidos desde {info['ips']}")

# Actividad por hora
print("\nActividad por hora:")
por_hora = analizador.actividad_por_hora()
for hora, cantidad in sorted(por_hora.items()):
    barra = '‚ñà' * (cantidad // 2)
    print(f"  {hora:02d}:00 {barra} ({cantidad})")

# Filtrar errores
print("\nTodos los errores:")
errores = analizador.filtrar(nivel='ERROR')
for log in errores:
    print(f"  [{log.fecha.strftime('%H:%M')}] {log.usuario} - {log.accion}")

# Generar reportes
print("\nGenerando reportes...")
analizador.generar_reporte('reporte_logs.txt')
analizador.exportar_json('logs_analisis.json')
print("‚úÖ Reportes generados")
````
