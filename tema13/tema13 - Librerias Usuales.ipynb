## 1. NumPy - Computación Numérica

**NumPy** (Numerical Python) es la librería fundamental para computación científica en Python.

### ¿Por qué NumPy?

### 1.1. Arrays de NumPy (ndarray)

````python
import numpy as np

# Crear arrays de diferentes formas
# Array 1D desde una lista
arr1d = np.array([1, 2, 3, 4, 5])
print("Array 1D:")
print(arr1d)
print(f"Tipo: {type(arr1d)}")
print(f"Forma: {arr1d.shape}")
print(f"Dimensiones: {arr1d.ndim}")
print(f"Tamaño: {arr1d.size}")
print(f"Tipo de datos: {arr1d.dtype}")
````

### 1.2. Funciones de Creación de Arrays

````python
# Arrays de ceros
zeros = np.zeros((3, 4))
print("Array de ceros (3x4):")
print(zeros)

# Arrays de unos
ones = np.ones((2, 3, 4))
print("\nArray de unos (2x3x4):")
print(ones)

# Array con un valor específico
full = np.full((2, 3), 7)
print("\nArray lleno de 7s (2x3):")
print(full)
````

### 1.3. Indexación y Slicing

````python
# Crear un array para ejemplos
arr = np.arange(1, 26).reshape(5, 5)
print("Array original:")
print(arr)

# Indexación básica
print(f"\nElemento [0, 0]: {arr[0, 0]}")
print(f"Elemento [2, 3]: {arr[2, 3]}")
print(f"Última fila: {arr[-1]}")
print(f"Última columna: {arr[:, -1]}")
````

````python
# Slicing
print("Primera fila:")
print(arr[0, :])

print("\nPrimera columna:")
print(arr[:, 0])

print("\nSubmatriz (filas 1-3, columnas 2-4):")
print(arr[1:4, 2:5])

print("\nFilas pares:")
print(arr[::2, :])
````

````python
# Indexación booleana
print("Elementos mayores que 15:")
print(arr[arr > 15])

print("\nElementos pares:")
print(arr[arr % 2 == 0])

# Modificar elementos con indexación booleana
arr_copia = arr.copy()
arr_copia[arr_copia > 20] = 0
print("\nArray con valores >20 cambiados a 0:")
print(arr_copia)
````

### 1.4. Operaciones Matemáticas

````python
# Operaciones elemento a elemento
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

print("a:", a)
print("b:", b)
print(f"\na + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
print(f"b / a = {b / a}")
print(f"a ** 2 = {a ** 2}")
````

### 1.6. Reshape y Manipulación de Forma

### 1.7. Broadcasting

````python
# Broadcasting permite operaciones entre arrays de diferentes formas
# Ejemplo 1: Escalar con array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print("Array:")
print(arr)

print("\nArray + 10:")
print(arr + 10)

# Ejemplo 2: Array 1D con array 2D
vector = np.array([10, 20, 30])
print("\nVector:")
print(vector)

print("\nArray + Vector:")
print(arr + vector)
````

````python
# Broadcasting más complejo
matriz_a = np.arange(1, 13).reshape(4, 3)
vector_b = np.array([100, 200, 300])

print("Matriz A (4x3):")
print(matriz_a)

print("\nVector B (1x3):")
print(vector_b)

print("\nMatriz A + Vector B:")
print(matriz_a + vector_b)
````

### 1.8. Álgebra Lineal

````python
# Producto punto (dot product)
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Vector a:", a)
print("Vector b:", b)
print(f"Producto punto: {np.dot(a, b)}")
print(f"Usando @: {a @ b}")

# Multiplicación de matrices
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print("\nMatriz A:")
print(A)
print("\nMatriz B:")
print(B)
print("\nA @ B:")
print(A @ B)
````

````python
# Otras operaciones de álgebra lineal
matriz = np.array([[1, 2],
                   [3, 4]])

print("Matriz:")
print(matriz)

print(f"\nDeterminante: {np.linalg.det(matriz):.2f}")

print("\nMatriz inversa:")
print(np.linalg.inv(matriz))

print("\nAutovalores y autovectores:")
eigenvalues, eigenvectors = np.linalg.eig(matriz)
print("Autovalores:", eigenvalues)
print("Autovectores:")
print(eigenvectors)
````

### 1.9. Ejemplo Práctico: Análisis de Notas

````python
# Simular notas de 30 estudiantes en 5 asignaturas
np.random.seed(42)
notas = np.random.randint(0, 101, size=(30, 5))

asignaturas = ['Matemáticas', 'Física', 'Química', 'Programación', 'Inglés']

print("Primeros 5 estudiantes:")
print(notas[:5])

# Análisis por asignatura
print("\n=== ESTADÍSTICAS POR ASIGNATURA ===")
for i, asig in enumerate(asignaturas):
    notas_asig = notas[:, i]
    print(f"\n{asig}:")
    print(f"  Media: {np.mean(notas_asig):.2f}")
    print(f"  Mediana: {np.median(notas_asig):.2f}")
    print(f"  Desv. Est.: {np.std(notas_asig):.2f}")
    print(f"  Mín: {np.min(notas_asig)}, Máx: {np.max(notas_asig)}")

# Análisis por estudiante
promedios = np.mean(notas, axis=1)
print("\n=== MEJORES ESTUDIANTES (Top 5) ===")
top_indices = np.argsort(promedios)[-5:][::-1]
for i, idx in enumerate(top_indices, 1):
    print(f"{i}. Estudiante {idx+1}: {promedios[idx]:.2f}")

# Estudiantes aprobados (promedio >= 50)
aprobados = np.sum(promedios >= 50)
print(f"\nEstudiantes aprobados: {aprobados}/{len(promedios)}")
print(f"Porcentaje: {(aprobados/len(promedios)*100):.1f}%")
````

### 2.5. Manejo de Datos Faltantes

### 2.6. Agrupación y Agregación (GroupBy)

````python
# Crear dataset de ventas
ventas = pd.DataFrame({
    'Producto': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'Vendedor': ['Juan', 'Ana', 'Pedro', 'Juan', 'Ana', 'Juan', 'Pedro', 'Pedro'],
    'Cantidad': [10, 15, 8, 12, 20, 14, 9, 11],
    'Precio': [100, 150, 100, 200, 150, 100, 200, 150]
})

ventas['Total'] = ventas['Cantidad'] * ventas['Precio']

print("Ventas:")
print(ventas)

# Agrupar por producto
print("\nVentas totales por producto:")
print(ventas.groupby('Producto')['Total'].sum())
````

````python
# Múltiples agregaciones
print("Estadísticas por producto:")
print(ventas.groupby('Producto')['Total'].agg(['sum', 'mean', 'count']))

# Agrupar por múltiples columnas
print("\nVentas por vendedor y producto:")
print(ventas.groupby(['Vendedor', 'Producto'])['Total'].sum())
````

````python
# Pivot table
print("Tabla pivote (Vendedor vs Producto):")
pivot = ventas.pivot_table(
    values='Total',
    index='Vendedor',
    columns='Producto',
    aggfunc='sum',
    fill_value=0
)
print(pivot)
````

### 2.8. Merge y Join - Combinar DataFrames

````python
# Crear dos DataFrames relacionados
empleados = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Nombre': ['Ana', 'Juan', 'María', 'Carlos'],
    'Departamento_ID': [1, 2, 1, 3]
})

departamentos = pd.DataFrame({
    'Departamento_ID': [1, 2, 3],
    'Departamento': ['Ventas', 'IT', 'RRHH']
})

print("Empleados:")
print(empleados)
print("\nDepartamentos:")
print(departamentos)

# Merge (join)
resultado = pd.merge(empleados, departamentos, on='Departamento_ID')
print("\nMerge (Inner Join):")
print(resultado)
````

````python
# Diferentes tipos de join
# Left join
left_join = pd.merge(empleados, departamentos, on='Departamento_ID', how='left')
print("Left Join:")
print(left_join)

# Concatenar verticalmente
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
concatenado = pd.concat([df1, df2], ignore_index=True)
print("\nConcatenación vertical:")
print(concatenado)
````

### 2.9. Ejemplo Práctico: Análisis de Ventas

````python
# Crear dataset de ventas realista
np.random.seed(42)

fechas = pd.date_range('2024-01-01', periods=100, freq='D')
productos = ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Webcam']
regiones = ['Norte', 'Sur', 'Este', 'Oeste']

ventas_data = pd.DataFrame({
    'Fecha': np.random.choice(fechas, 200),
    'Producto': np.random.choice(productos, 200),
    'Region': np.random.choice(regiones, 200),
    'Cantidad': np.random.randint(1, 20, 200),
    'Precio_Unitario': np.random.randint(10, 1000, 200)
})

ventas_data['Total'] = ventas_data['Cantidad'] * ventas_data['Precio_Unitario']
ventas_data['Mes'] = ventas_data['Fecha'].dt.month

print("Dataset de ventas (primeras 10 filas):")
print(ventas_data.head(10))

print("\n=== ANÁLISIS DE VENTAS ===")
print(f"Total de transacciones: {len(ventas_data)}")
print(f"Ingresos totales: ${ventas_data['Total'].sum():,.2f}")
print(f"Ticket promedio: ${ventas_data['Total'].mean():,.2f}")
````

````python
# Análisis por producto
print("\n=== VENTAS POR PRODUCTO ===")
por_producto = ventas_data.groupby('Producto').agg({
    'Total': ['sum', 'mean', 'count'],
    'Cantidad': 'sum'
}).round(2)
print(por_producto)

# Análisis por región
print("\n=== VENTAS POR REGIÓN ===")
por_region = ventas_data.groupby('Region')['Total'].agg([
    ('Total_Ventas', 'sum'),
    ('Promedio', 'mean'),
    ('Transacciones', 'count')
]).round(2)
print(por_region)
````

````python
# Top 5 días con más ventas
print("\n=== TOP 5 DÍAS CON MÁS VENTAS ===")
ventas_por_dia = ventas_data.groupby('Fecha')['Total'].sum().sort_values(ascending=False)
print(ventas_por_dia.head())

# Producto más vendido por región
print("\n=== PRODUCTO MÁS VENDIDO POR REGIÓN ===")
for region in regiones:
    datos_region = ventas_data[ventas_data['Region'] == region]
    top_producto = datos_region.groupby('Producto')['Total'].sum().idxmax()
    total = datos_region.groupby('Producto')['Total'].sum().max()
    print(f"{region}: {top_producto} (${total:,.2f})")
````

### 3.1. Gráficos de Líneas

````python
import matplotlib.pyplot as plt

# Gráfico de línea simple
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Función Seno')
plt.xlabel('X')
plt.ylabel('sen(X)')
plt.grid(True)
plt.show()

print("Gráfico creado")
````

## 4. Seaborn - Visualización Estadística

**Seaborn** es una librería de visualización basada en Matplotlib que ofrece gráficos estadísticos de alto nivel.

### ¿Por qué Seaborn?
- **Estética:** Gráficos atractivos con temas predefinidos
- **Integración:** Funciona directamente con DataFrames de Pandas
- **Estadística:** Gráficos estadísticos complejos simplificados
- **Paletas:** Esquemas de colores profesionales
- **Menos código:** Resultados impresionantes con pocas líneas

### 4.1. Configuración y Temas

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo de Seaborn
sns.set_theme(style="whitegrid")

# Crear datos de ejemplo
tips = sns.load_dataset('tips')  # Dataset integrado
print("Dataset 'tips' (primeras 10 filas):")
print(tips.head(10))

print(f"\nForma del dataset: {tips.shape}")
print(f"Columnas: {tips.columns.tolist()}")
````

````python
# Diferentes estilos de Seaborn
estilos = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for i, estilo in enumerate(estilos):
    sns.set_style(estilo)
    ax = axes[i]
    sns.scatterplot(data=tips, x='total_bill', y='tip', ax=ax)
    ax.set_title(f'Estilo: {estilo}')

axes[5].axis('off')
plt.tight_layout()
plt.show()

# Restaurar estilo por defecto
sns.set_theme()
````

### 4.2. Gráficos de Distribución

````python
# Histograma con curva de densidad
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(data=tips, x='total_bill', kde=True, color='steelblue')
plt.title('Distribución de Cuentas Totales')

plt.subplot(1, 2, 2)
sns.histplot(data=tips, x='tip', kde=True, color='coral')
plt.title('Distribución de Propinas')

plt.tight_layout()
plt.show()
````

````python
# Box plots (diagramas de caja)
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=tips, hue='day', y='total_bill', palette='Set2')
plt.title('Cuentas por Día de la Semana')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.boxplot(data=tips, x='time', y='tip', hue='sex', palette='Set1')
plt.title('Propinas por Momento del Día y Género')

plt.tight_layout()
plt.show()
````

````python
# Violin plots (combinan boxplot y densidad)
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.violinplot(data=tips, hue='day', y='total_bill', palette='muted')
plt.title('Distribución de Cuentas por Día (Violin Plot)')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', 
               split=True, palette='pastel')
plt.title('Distribución de Cuentas por Día y Género (Split)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````

### 4.3. Gráficos de Relación

````python
# Scatter plot con regresión lineal
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', 
                size='size', sizes=(50, 200), alpha=0.7)
plt.title('Relación entre Cuenta Total y Propina')

plt.subplot(1, 2, 2)
sns.regplot(data=tips, x='total_bill', y='tip', 
            scatter_kws={'alpha':0.5}, color='darkblue')
plt.title('Regresión Lineal: Cuenta vs Propina')

plt.tight_layout()
plt.show()

# Calcular correlación
correlacion = tips['total_bill'].corr(tips['tip'])
print(f"Correlación: {correlacion:.3f}")
````

````python
# Pair plot - visualizar todas las relaciones
# Seleccionar solo variables numéricas
sns.pairplot(tips, hue='time', palette='husl', 
             vars=['total_bill', 'tip', 'size'],
             height=2.5)
plt.suptitle('Pair Plot - Relaciones entre Variables', y=1.02)
plt.show()
````

### 4.4. Mapas de Calor (Heatmaps)

````python
# Matriz de correlación
# Seleccionar solo columnas numéricas
numeric_cols = tips.select_dtypes(include=[np.number]).columns
correlacion = tips[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlacion, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Matriz de Correlación', fontsize=16, fontweight='bold')
plt.show()
````

````python
# Heatmap de tabla pivote
pivot_data = tips.pivot_table(values='tip', index='day', 
                               columns='time', aggfunc='mean', observed=False)

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_data, annot=True, fmt='.2f', cmap='YlGnBu',
            linewidths=0.5, cbar_kws={'label': 'Propina Promedio'})
plt.title('Propina Promedio por Día y Momento', fontsize=14, fontweight='bold')
plt.xlabel('Momento del Día')
plt.ylabel('Día de la Semana')
plt.show()
````

### 4.5. Gráficos Categóricos

````python
# Count plot (gráfico de conteo)
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.countplot(data=tips, x='day', hue='sex', palette='Set2')
plt.title('Número de Clientes por Día y Género')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.countplot(data=tips, x='size', palette='viridis', hue='size')
plt.title('Distribución del Tamaño de Grupos')

plt.tight_layout()
plt.show()
````

````python
# Bar plot con valores agregados
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.barplot(data=tips, hue='day', y='total_bill', 
            estimator=np.mean, errorbar=('ci', 95), palette='rocket')
plt.title('Cuenta Promedio por Día (con IC 95%)')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.barplot(data=tips, x='time', y='tip', hue='sex', 
            estimator=np.median, palette='mako')
plt.title('Propina Mediana por Momento y Género')

plt.tight_layout()
plt.show()
````

### 4.6. FacetGrid - Gráficos Múltiples

````python
# FacetGrid para visualizaciones por categorías
g = sns.FacetGrid(tips, col='time', row='sex', height=4, aspect=1.2)
g.map(sns.scatterplot, 'total_bill', 'tip', alpha=0.7)
g.add_legend()
g.fig.suptitle('Relación Cuenta-Propina por Género y Momento', 
               y=1.02, fontsize=14, fontweight='bold')
plt.show()
````

````python
# FacetGrid con histogramas
g = sns.FacetGrid(tips, col='day', col_wrap=2, height=4)
g.map(sns.histplot, 'total_bill', kde=True, color='steelblue')
g.fig.suptitle('Distribución de Cuentas por Día', 
               y=1.02, fontsize=14, fontweight='bold')
plt.show()
````

### 4.7. Ejemplo Práctico: Dashboard Estadístico

````python
# Dashboard completo con Seaborn
fig = plt.figure(figsize=(18, 12))

# 1. Distribución de propinas
ax1 = plt.subplot(3, 3, 1)
sns.histplot(data=tips, x='tip', kde=True, color='coral', ax=ax1)
ax1.set_title('Distribución de Propinas', fontweight='bold')

# 2. Boxplot por día
ax2 = plt.subplot(3, 3, 2)
sns.boxplot(data=tips, hue='day', y='total_bill', palette='Set2', ax=ax2)
ax2.set_title('Cuentas por Día', fontweight='bold')
plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45)

# 3. Count plot de clientes
ax3 = plt.subplot(3, 3, 3)
sns.countplot(data=tips, x='size', palette='viridis', ax=ax3, hue='size')
ax3.set_title('Tamaño de Grupos', fontweight='bold')

# 4. Scatter con regresión
ax4 = plt.subplot(3, 3, 4)
sns.regplot(data=tips, x='total_bill', y='tip', ax=ax4, color='darkblue')
ax4.set_title('Cuenta vs Propina', fontweight='bold')

# 5. Violin plot
ax5 = plt.subplot(3, 3, 5)
sns.violinplot(data=tips, hue='time', y='tip', palette='muted', ax=ax5)
ax5.set_title('Propinas por Momento', fontweight='bold')

# 6. Bar plot por género
ax6 = plt.subplot(3, 3, 6)
sns.barplot(data=tips, hue='sex', y='total_bill', 
            estimator=np.mean, palette='pastel', ax=ax6)
ax6.set_title('Cuenta Promedio por Género', fontweight='bold')

# 7. Heatmap de correlación
ax7 = plt.subplot(3, 3, 7)
corr_matrix = tips[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, ax=ax7,
            square=True, cbar_kws={"shrink": 0.8})
ax7.set_title('Matriz de Correlación', fontweight='bold')

# 8. Distribución por día y género
ax8 = plt.subplot(3, 3, 8)
tips_grouped = tips.groupby(['day', 'sex'], observed=False).size().reset_index(name='count')
pivot = tips_grouped.pivot(index='day', columns='sex', values='count')
pivot.plot(kind='bar', ax=ax8, color=['lightcoral', 'lightblue'])
ax8.set_title('Clientes por Día y Género', fontweight='bold')
plt.setp(ax8.xaxis.get_majorticklabels(), rotation=45)

# 9. Scatter categorizado
ax9 = plt.subplot(3, 3, 9)
sns.scatterplot(data=tips, x='total_bill', y='tip', 
                hue='time', size='size', sizes=(50, 200), 
                alpha=0.7, ax=ax9)
ax9.set_title('Cuenta vs Propina (Múltiples Variables)', fontweight='bold')

plt.suptitle('Dashboard Estadístico Completo - Restaurante', 
             fontsize=18, fontweight='bold', y=0.995)
plt.tight_layout()
plt.show()
````

## 📚 Resumen

### NumPy
- **Arrays:** Estructuras de datos eficientes para computación numérica
- **Operaciones:** Vectorizadas, mucho más rápidas que bucles Python
- **Broadcasting:** Operaciones entre arrays de diferentes formas
- **Álgebra Lineal:** Operaciones matriciales, determinantes, autovalores

### Pandas
- **Series y DataFrames:** Estructuras para datos tabulares
- **Lectura/Escritura:** CSV, Excel, JSON, SQL
- **Manipulación:** Filtrado, agrupación, merge
- **Datos Faltantes:** Detección y manejo de NaN

### Matplotlib
- **Gráficos:** Líneas, barras, dispersión, histogramas, circulares
- **Personalización:** Control total sobre estilos y elementos
- **Subplots:** Múltiples gráficos en una figura
- **Exportación:** PNG, PDF, SVG

### Seaborn
- **Visualización estadística:** Gráficos hermosos con menos código
- **Integración Pandas:** Trabaja directamente con DataFrames
- **Tipos de gráficos:** Distribución, relación, categóricos, matrices
- **FacetGrid:** Múltiples visualizaciones por categorías
